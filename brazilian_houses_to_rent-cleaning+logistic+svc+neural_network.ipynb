```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('../input/brasilian-houses-to-rent/houses_to_rent.csv')
df.head()
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['floor'].replace(to_replace='-',value=0,inplace=True)
```

```python
df.head()
```

```python
df['animal'].replace(to_replace='acept',value=1,inplace=True)
df['animal'].replace(to_replace='not acept',value=0,inplace=True)

df['furniture'].replace(to_replace='furnished',value=1,inplace=True)
df['furniture'].replace(to_replace='not furnished',value=0,inplace=True)
```

```python
df.head()
```

```python
for i in ['hoa','rent amount','property tax','fire insurance','total']:
    df[i].replace(to_replace='R\$',value='',regex=True,inplace=True)
    df[i].replace(to_replace=',',value='',regex=True,inplace=True)
```

```python
df.head()
```

```python
#casting entire df into integers
df=df.astype(np.int64)
```

```python
df.isin(['Sem info']).any()
```

```python
df['hoa'].replace(to_replace='Sem info',value=0,inplace=True)
```

```python
df.isin(['Sem info']).any()
```

```python
#casting entire df into integers
df=df.astype(np.int64)
```

```python
df.isin(['Incluso']).any()
```

```python
df['hoa'].replace(to_replace='Incluso',value=0,inplace=True)
df['property tax'].replace(to_replace='Incluso',value=0,inplace=True)
```

```python
df.isin(['Incluso']).any()
```

```python
df=df.astype(np.int64)
```

```python
df=df.sample(frac=1).reset_index(drop=True)
```

```python
df.head()
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.head()
```

```python
y=df['city']
x=df.drop(['city'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(x)
x=scaler.transform(x)
```

```python
x
```

```python
pd.DataFrame(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
logmodel=LogisticRegression(penalty='l2',verbose=1)
svmmodel=SVC(kernel='rbf',verbose=1)
nnmodel=MLPClassifier(hidden_layer_sizes=(16,16),activation='relu',solver='adam',verbose=1)
```

```python
logmodel.fit(x_train,y_train)
svmmodel.fit(x_train,y_train)
nnmodel.fit(x_train,y_train)
```

```python
print(logmodel.score(x_test,y_test))
print(svmmodel.score(x_test,y_test))
print(nnmodel.score(x_test,y_test))
```

```python
log_pred=logmodel.predict(x_test)
svm_pred=svmmodel.predict(x_test)
nn_pred=nnmodel.predict(x_test)
```
